### Week 7 and 8 Scatter Plot, Bubble Charts and Density Plot

```python
# Importing Libraries 

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# Load Data Life Expectancy 
df_crime = pd.read_csv("crimerates-by-state-2005.csv")
```

```python
df_crime.tail()
```

```python
df2= df_crime.loc[df_crime['state'] != "United States"]
```

```python
# Scatter Plot 
plt.plot( 'robbery', 'murder', data=df2, linestyle='none', marker='o')
plt.title("Robbery Rate against Murder Rate in Us in 2005")
plt.xlabel("Robbery Rate")
plt.ylabel("Murder Rate")

plt.show()
```

```python
# Bubble Charts
fig = plt.figure(figsize = (14, 8))
ax = sns.scatterplot(data=df2, x="burglary", y="murder", size="population", alpha=0.9,sizes=(20, 500))
ax.set(title='Murder Rate and Bulgary Rate by State Population Size in 2005 ')
plt.show()
```

```python
# Density Plot
fig = plt.figure(figsize = (14, 8))
ax2 = sns.kdeplot(df_crime['murder'],shade = True)
ax2.set(title='Murder Rate Distribution in US in 2005 ')
plt.show()
```
